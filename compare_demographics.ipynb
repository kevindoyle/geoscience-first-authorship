```python
# load results of the analysis.ipynb notebook
journ_dem = pd.read_csv("analysis_output_2021-06-09.csv")
journ_dem = journ_dem[journ_dem.year >= 2016]
dem = pd.DataFrame([journ_dem.groupby("year").P_first_F_year.mean()])
dem = dem.melt()
dem["p_female_firstauth"] = round(dem["value"] * 100., 2)
del dem["value"]

dem2 = pd.DataFrame([journ_dem.groupby("year").P_last_F_year.mean()])
dem2 = dem2.melt()

dem["p_female_lastauth"] = round(dem2["value"] * 100., 2)

#dem["p_female_egu_ecs"] = egu_dem["p_female_ecs"]
dem["p_female_egu_all"] = egu_dem["p_female_all"]

display(dem)
dem.set_index("year").plot.bar()
plt.ylabel("Frequency %")
plt.legend(["P_1st_f", "P_last_f", "P_EGU_f"])
plt.xlim([-1, 6])
plt.savefig("EGU_demograph_vs_author_16-20.png", dpi=150)
#plt.show()
```
